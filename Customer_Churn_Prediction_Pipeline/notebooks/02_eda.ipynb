### 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/Churn_Modelling.csv")
```

### 2. Churn Distribution

Business Question

How many customers leave compared to those who stay?

```python
sns.countplot(data=df, x="Exited")
plt.title("Customer Churn Distribution")
plt.show()
```

### 3. Churn by Gender

Business Question

Do males and females churn at different rates?

```python
sns.countplot(data=df, x="Gender", hue="Exited")
plt.title("Churn by Gender")
plt.show()
```

### 4. Churn by Geography

Business Question

Which country/region loses the most customers?

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Geography", hue="Exited")
plt.title("Churn by Geography")
plt.show()
```

### 5. Age vs Churn

Business Question

Are older customers more likely to leave?

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Exited", y="Age")
plt.title("Age Distribution by Churn")
plt.show()
```

### 6. Balance vs Churn

Business Question

Do customers with larger balances churn more often?

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Exited", y="Balance")
plt.title("Balance Distribution by Churn")
plt.show()
```

### 7. Active Member vs Churn

Business Question

Does customer engagement affect churn?

```python
sns.countplot(data=df, x="IsActiveMember", hue="Exited")
plt.title("Active Membership vs Churn")
plt.show()
```

### 8. Products vs Churn

Business Question

How does product ownership influence churn?

```python
sns.countplot(data=df, x="NumOfProducts", hue="Exited")
plt.title("Products Owned vs Churn")
plt.show()
```

### 9. Correlation Heatmap

Before modeling, we want to know which numerical features are related.

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df.select_dtypes(include=["int64","float64"]).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Feature Correlation Matrix")
plt.show()
```
